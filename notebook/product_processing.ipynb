```python
import sys, os
import pandas as pd

sys.path.append('..')
dotenv_path = os.path.join('../.env')
```

```python
from src.db_connection import get_db_connection

# Establish the database connection
conn = get_db_connection()
   
# Get a cursor object
cursor = conn.cursor()

# Close the connection
# conn.close()
```

```python

# Execute the SQL query to select data from the user table
cursor.execute("SELECT * FROM products")

# Fetch all the rows from the result set
rows = cursor.fetchall()

# Create a pandas DataFrame from the result set
product_df = pd.DataFrame(rows, columns=[col[0] for col in cursor.description])
```

```python
# Execute the SQL query to select data from the user table
cursor.execute("SELECT * FROM product_names")

# Fetch all the rows from the result set
rows = cursor.fetchall()

# Create a pandas DataFrame from the result set
product_names_df = pd.DataFrame(rows, columns=[col[0] for col in cursor.description])
```

```python
# Execute the SQL query to select data from the user table
cursor.execute("SELECT * FROM categories")

# Fetch all the rows from the result set
rows = cursor.fetchall()

# Create a pandas DataFrame from the result set
categories_df = pd.DataFrame(rows, columns=[col[0] for col in cursor.description])
```

```python
missing_values_count = product_df.isnull().sum()
print(missing_values_count)
```

```python
total_rows = product_df.shape[0]
missing_values_percentage = (product_df.isnull().sum() / total_rows) * 100
print(missing_values_percentage)
```

```python
selected_columns_df = product_df[['id', 'vendor_id', 'short_description', 'long_description','tags', 'name_id']]
selected_columns_df
```

```python

total_rows = selected_columns_df.shape[0]
missing_values_percentage = (selected_columns_df.isnull().sum() / total_rows) * 100
print(missing_values_percentage)
```

```python
# Assuming product_name is a DataFrame containing the 'id' and 'name' columns

# Merge selected_columns_df with product_name based on the 'id' column
selected_with_names_df = selected_columns_df.merge(selected_name_df, on='name_id', how='left')

# Display the merged DataFrame
selected_with_names_df
```

```python
# Assuming product_name is a DataFrame containing the 'id' and 'name' columns

# Merge selected_columns_df with product_name based on the 'id' column
Merged_product_df = selected_with_names_df.merge(selected_category_df, on='category_id', how='left')

# Display the merged DataFrame
Merged_product_df
```

```python
Merged_product_df = Merged_product_df.rename(columns={'id': 'product_id', 'name_x': 'product_name', 'name_y': 'product_category'})
Merged_product_df
```

```python
# Save the cleaned DataFrame to a CSV file
# Ensure the 'CleanedData' directory exists
folder_path ="../data/CleanedData"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Specify the full path including the file name
file_path = os.path.join(folder_path, 'cleaned_products.csv')

# Save the DataFrame
Merged_product_df.to_csv(file_path, index=False)
```
