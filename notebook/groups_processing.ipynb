```python

# Establish the database connection
conn = get_db_connection()
   
# Get a cursor object
cursor = conn.cursor()

# Close the connection
# conn.close()
```

```python
# Define your SQL queries for each table
query_groups = "SELECT * FROM groups;"
query_group_carts = "SELECT * FROM groups_carts;"
query_group_deals = "SELECT * FROM group_deals;"

# Load data from each query into separate DataFrames
cursor.execute(query_groups)
groups_data = cursor.fetchall()
groups_df = pd.DataFrame(groups_data, columns=[col[0] for col in cursor.description])

cursor.execute(query_group_carts)
group_carts_data = cursor.fetchall()
group_carts_df = pd.DataFrame(group_carts_data, columns=[col[0] for col in cursor.description])

cursor.execute(query_group_deals)
group_deals_data = cursor.fetchall()
group_deals_df = pd.DataFrame(group_deals_data, columns=[col[0] for col in cursor.description])
```

```python
# Check for missing values in groups_df
groups_missing_values = groups_df.isnull().sum()

# Check for missing values in group_carts_df
group_carts_missing_values = group_carts_df.isnull().sum()

# Check for missing values in group_deals_df
group_deals_missing_values = group_deals_df.isnull().sum()

print("Missing values in groups_df:")
print(groups_missing_values)

print("\nMissing values in group_carts_df:")
print(group_carts_missing_values)

print("\nMissing values in group_deals_df:")
group_deals_missing_values
```

```python
# Remove the 'deleted_at' column from groups_df
groups_df = groups_df.drop(columns=['deleted_at','updated_at','status', 'available_for_delivery'])
# Remove the 'deleted_at' column from group_carts_df
group_carts_df = group_carts_df.drop(columns=['status','quantity','deleted_at','updated_at'])

# Remove the 'deleted_at' column from group_deals_df
group_deals_df = group_deals_df.drop(columns=['deleted_at', 'updated_at','max_group_member','group_price','effective_from','effective_to','expiration_time'])
```

### Combine groups_df with group_deals_df to link each group with specific product deals:

```python
#'group_deals_id' in groups_df refers to 'id' in group_deals_df:

group_info_df = pd.merge(groups_df, group_deals_df, left_on='group_deals_id', right_on='id', suffixes=('_group', '_deal'))
```

### Add User Information to Group Data


```python

#Link group_carts_df to the unified group dataframe to include user data
# Assuming 'group_id' in group_carts_df refers to 'id' in groups_df:
full_group_df = pd.merge(group_info_df, group_carts_df, left_on='id_group', right_on='group_id')
full_group_df
```

```python
full_group_df['formatted_date'] = pd.to_datetime(full_group_df['created_at_deal']).dt.strftime('%d/%m/%Y')
```

```python
full_group_df.drop(columns=['day_of_week','day_of_month'])
```

```python
# Save the cleaned DataFrame to a CSV file
# Ensure the 'CleanedData' directory exists
folder_path = "CleanedData"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Specify the full path including the file name
file_path = os.path.join(folder_path, 'cleaned_groups.csv')

# Save the DataFrame
full_group_df.to_csv(file_path, index=False)
```
