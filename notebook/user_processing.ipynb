```python
import sys, os
import pandas as pd

sys.path.append('..')
dotenv_path = os.path.join('../.env')
```

```python
from utils.db_connection import get_db_connection

# Establish the database connection
conn = get_db_connection()

# Get a cursor object
cursor = conn.cursor()

# Close the connection
# conn.close()
```

```python

# Execute the SQL query to select data from the user table
cursor.execute("SELECT * FROM users")

# Fetch all the rows from the result set
rows = cursor.fetchall()

# Create a pandas DataFrame from the result set
user_df = pd.DataFrame(rows, columns=[col[0] for col in cursor.description])
```

```python
missing_values_count = user_df.isnull().sum()
print(missing_values_count)
```

```python
total_rows = user_df.shape[0]
missing_values_percentage = (user_df.isnull().sum() / total_rows) * 100
print(missing_values_percentage)
```

```python
selected_columns_df = user_df[['id', 'gender', 'user_birthday', 'country_id']]
selected_columns_df
```

```python


# Assuming user_df is your original DataFrame

# Calculate the count of missing values for the specified columns
missing_values_count = selected_columns_df[['id', 'gender', 'country_id']].isnull().sum()

# Calculate the total number of rows in the DataFrame
total_rows = selected_columns_df.shape[0]

# Calculate the percentage of missing values for each specified column
missing_values_percentage = (selected_columns_df[['id',  'gender', 'country_id']].isnull().sum() / total_rows) * 100

# Create a new DataFrame to display missing values count and percentage side by side
missing_values_summary = pd.DataFrame({
    'Column': ['id', 'gender', 'country_id'],
    'Missing Values': missing_values_count.values,
    'Percentage (%)': missing_values_percentage.values
})

print(missing_values_summary)
```

```python
import numpy as np
def random_gender(value):
    if pd.isnull(value):
        return np.random.choice(['Male', 'Female'])
    else:
        return value

# Apply the function to the gender column to fill null values randomly
selected_columns_df['gender'] = selected_columns_df['gender'].apply(random_gender)

# Verify the changes
print(selected_columns_df['gender'].value_counts(dropna=False))
```

```python
import pandas as pd

# Assuming user_df is your original DataFrame

# Define a mapping of values to replace
gender_replacement = {
    'áˆ´á‰µ': 'Female',  # Replace with 'Female'
    'á‹ˆáŠ•á‹µ': 'Male',   # Replace with 'Male'
    'female': 'Female',  # Replace with 'Female'
    'male': 'Male',  # Replace with 'Male'
    'Femele': 'Female'  # Replace with 'Female'
}

# Use the replace() function to replace the values in the gender column
selected_columns_df['gender'] = selected_columns_df['gender'].replace(gender_replacement)

# Verify the changes
print(selected_columns_df['gender'].value_counts())
# selected_columns_df
```
