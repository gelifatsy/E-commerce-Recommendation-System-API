```python
import sys, os
import pandas as pd

sys.path.append('..')
dotenv_path = os.path.join('../.env')
```

```python
from utils.db_connection import get_db_connection

# Establish the database connection
conn = get_db_connection()

# Get a cursor object
cursor = conn.cursor()

# Close the connection
# conn.close()
```

```python

# Execute the SQL query to select data from the user table
cursor.execute("SELECT * FROM personal_cart_items")

# Fetch all the rows from the result set
rows = cursor.fetchall()

# Create a pandas DataFrame from the result set
personal_cart_items_df = pd.DataFrame(rows, columns=[col[0] for col in cursor.description])
```

```python
personal_cart_items_df
```

```python
missing_values_count = personal_cart_items_df.isnull().sum()
print(missing_values_count)
```

```python
personal_cart_items_df.info()
```

```python
total_rows = personal_cart_items_df.shape[0]
missing_values_percentage = (personal_cart_items_df.isnull().sum() / total_rows) * 100
print(missing_values_percentage)
```

```python
selected_columns_df = personal_cart_items_df[['id', 'cart_id', 'product_id']]
selected_columns_df
```

```python

total_rows = selected_columns_df.shape[0]
missing_values_percentage = (selected_columns_df.isnull().sum() / total_rows) * 100
print(missing_values_percentage)
```

```python
# Save the cleaned DataFrame to a CSV file
# Ensure the 'CleanedData' directory exists
folder_path = "CleanedData"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Specify the full path including the file name
file_path = os.path.join(folder_path, 'cleaned_personal_cart.csv')

# Save the DataFrame
selected_columns_df.to_csv(file_path, index=False)
```
